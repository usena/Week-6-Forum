```python
data={
    "lloyd" : {
        "name": "Lloyd",
        "homework": [90.0,97.0,75.0,92.0],
        "quizzes": [88.0,40.0,94.0],
        "tests": [75.0,90.0]
    },
    "alice" : {
        "name": "Alice",
        "homework": [100.0, 92.0, 98.0, 100.0],
        "quizzes": [82.0, 83.0, 91.0],
        "tests": [89.0, 97.0]
    },
    "tyler" : {
        "name": "Tyler",
        "homework": [0.0, 87.0, 75.0, 22.0],
        "quizzes": [0.0, 75.0, 78.0],
        "tests": [100.0, 100.0]
    }
}

students=["lloyd", "alice", "tyler"]

for x in students:
    print(data[x].get("name"))
    print(f'Homework scores:{data[x].get("homework")}')
    print(f'Quiz scores:{data[x].get("quizzes")}')
    print(f'Test scores:{data[x].get("tests")}')

def average(numbers):
    total=float(sum(numbers))
    total=total/len(numbers)
    return total

def get_average(student):
    homework=average(data[student].get("homework"))*0.1
    quizzes=average(data[student].get("quizzes"))*0.3
    tests=average(data[student].get("tests"))*0.6
    get_total=homework+quizzes+tests
    return get_total

def get_letter_grade(score):
    if score >= 90: 
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "F"

def get_class_average(students):
    result=[]
    for y in students:
        result.append(get_average(y))
    result=average(result)
    return result

print(f'Class Average: {get_class_average(students)}')
print(f'Class Letter Grade: {get_letter_grade(get_class_average(students))}')
```
